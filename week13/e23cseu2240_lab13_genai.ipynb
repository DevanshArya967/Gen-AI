```python
import random
import numpy as np

# Dataset
sentences = [
    "AI is transforming industries",
    "Machine learning enables automation",
    "Deep learning models generate data",
    "Reinforcement learning improves decisions"
]

# Initialize probabilities equally
probabilities = np.ones(len(sentences)) / len(sentences)

# Reward function
def get_reward(sentence):
    if sentence in sentences:
        return 1
    return 0

# Generate sentence based on probabilities
def generate_sentence():
    return np.random.choice(sentences, p=probabilities)

# Reinforcement Learning training
learning_rate = 0.1

print("Before RL Training:\n")
for i in range(5):
    print(generate_sentence())

# Training loop
for episode in range(20):
    generated = generate_sentence()
    reward = get_reward(generated)

    index = sentences.index(generated)

    # Update probability using reward
    probabilities[index] += learning_rate * reward

    # Normalize probabilities
    probabilities[:] = probabilities / probabilities.sum()

# After training
print("\nAfter RL Training:\n")
for i in range(10):
    print(generate_sentence())

# Final probabilities
print("\nUpdated Sentence Probabilities:")
for s, p in zip(sentences, probabilities):
    print(f"{s} --> {p:.3f}")
```
